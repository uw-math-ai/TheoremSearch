```python
import json
from pathlib import Path
from slogans import generate_theorem_slogans
from embeddings import embed_texts

PARSED_PAPERS_DIR = Path("../parsed_papers")

N = len(list(PARSED_PAPERS_DIR.iterdir()))
```

```python
for j, parsed_paper_path in enumerate(list(PARSED_PAPERS_DIR.iterdir())):
    print(str(parsed_paper_path))

    with open(parsed_paper_path, "r") as f:
        parsed_paper = json.load(f)

    if parsed_paper["theorem_embeddings"][0]["theorem_slogan"] is None:
        theorems = [theorem["theorem_body"] for theorem in parsed_paper["theorem_embeddings"]]

        print(" > Generating slogans", end=" ")
        slogans = generate_theorem_slogans(theorems, "")

        for i, slogan in enumerate(slogans):
            parsed_paper["theorem_embeddings"][i]["theorem_slogan"] = slogan
    else:
        slogans = [theorem_embedding["theorem_slogan"] for theorem_embedding in parsed_paper["theorem_embeddings"]]

    if parsed_paper["theorem_embeddings"][0]["embedding"] is None:
        print(" > Embedding texts")
        embeddings = embed_texts(slogans)

        for i, embedding in enumerate(embeddings):
            parsed_paper["theorem_embeddings"][i]["embedding"] = embedding

    with open(parsed_paper_path, "w") as f:
        json.dump(parsed_paper, f, indent=4)

    print(f" > Progress {j+1}/{N}")
```

```python
import json
from pathlib import Path
from rds import get_rds_connection, upload_theorem_metadata_and_embeddings

PARSED_PAPERS_DIR = Path("../parsed_papers")

conn = get_rds_connection()

for i, parsed_paper_path in enumerate(PARSED_PAPERS_DIR.iterdir()):
    with open(parsed_paper_path, "r") as f:
        parsed_paper = json.load(f)

    upload_theorem_metadata_and_embeddings(conn, parsed_paper["theorem_metadata"], parsed_paper["theorem_embeddings"])

    print(f"Uploaded: {i+1}/{N}")
```
